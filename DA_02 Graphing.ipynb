```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("datasets/avocado.csv")
albany_df = df[df["region"] == "Albany"]
albany_df.set_index("Date",inplace = True)
albany_df["AveragePrice"].plot()
```

```python
"""When a csv file is imported and a Data Frame is made, the Date time objects in 
the file are read as a string object rather a Date Time object and Hence it’s very 
tough to perform operations like Time difference on a string rather a Date Time object.
Pandas to_datetime() method helps to convert string Date time into Python Date time object.
"""
```

```python
#in above graph dates are running over each other this means that does not realize that these are dates
df['Date'] = pd.to_datetime(df["Date"])
albany_df["AveragePrice"].plot()
```

```python
#to make sure that dates are in order
albany_df.sort_index(inplace = True)
```

```python
#rolling means that we take a window of size k at a time and perform some operations on it
albany_df['AveragePrice'].rolling(25).mean().plot()
```

```python
albany_df['price25ma'] = albany_df['AveragePrice'].rolling(25).mean()
```

```python
#using this copy the dataframe so that it does not affect the original one and also you wont get warning now
albany_df = df.copy()[df["region"] == 'Albany']
```

```python
#we want to restructure the dataframe so that the columns are ther regions and 
#the values of those columns are price25ma
#all of these have same column name so we are gonna use f strings
graph_df = pd.DataFrame()
for region in df['region'].unique()[:16]:
    print(region)
    region_df = df.copy()[df['region'] == region]
    #albany_df.set_index("Date",inplace = True)
    #albany_df.sort_index(inplace = True)
    albany_df[f'{region}_price25ma'] = region_df['AveragePrice'].rolling(25).mean()  
    
    if graph_df.empty:
        graph_df = region_df[[f"{region}_price25ma"]]
    else:
        #if you have 2 dataframe indexed in same way then you call both and they will be joined for each index
        graph_df = graph_df.join(region_df[f"{region}_price25ma"])
```
